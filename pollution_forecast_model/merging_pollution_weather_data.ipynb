```python
import pandas as pd
```

```python
weather_df = pd.read_csv('Data/weather_data_skopje_2022_to_2024.csv')
```

```python
df_pm10 = pd.read_csv('Data/pm10_data_cleaned.csv')
df_pm2_5 = pd.read_csv('Data/pm2_5_data_cleaned.csv')
df_no2 = pd.read_csv('Data/no2_data_cleaned.csv')
```

```python
df_pm10['timestamp'] = pd.to_datetime(df_pm10[['year', 'month', 'day', 'hour']].apply(lambda x: x.astype(str).str.zfill(2)).agg('-'.join, axis=1))
```

```python
df_pm2_5['timestamp'] = pd.to_datetime(df_pm2_5[['year', 'month', 'day', 'hour']].apply(lambda x: x.astype(str).str.zfill(2)).agg('-'.join, axis=1))
df_no2['timestamp'] = pd.to_datetime(df_no2[['year', 'month', 'day', 'hour']].apply(lambda x: x.astype(str).str.zfill(2)).agg('-'.join, axis=1))
```

```python
weather_df['timestamp'] = pd.to_datetime(weather_df['stamp'])
```

```python
weather_df.drop(columns=['stamp'], inplace=True)
```

```python
weather_df = weather_df[weather_df['timestamp'].dt.hour != 0]
```

```python
weather_df
```

```python
df_pm10_merged = pd.merge(df_pm10, weather_df, on='timestamp', how='left')
df_pm2_5_merged = pd.merge(df_pm2_5, weather_df, on='timestamp', how='left')
df_no2_merged = pd.merge(df_no2, weather_df, on='timestamp', how='left')
```
